```python
# Introduction: Iris
# This exercise may seem a little bit strange, but keep doing it.
# Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# import dataset and Assign it to a variable called iris
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url)

iris.head()
```

```python
# Create columns

iris.columns = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class']
iris.head()
```

```python
# Is there any missing value in the dataframe?
pd.isnull(iris).sum()
```

```python
# Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

iris.iloc[10:30,2:3] = np.nan
iris.head(20)
```

```python
# Good, now lets substitute the NaN values to 1.0
iris.petal_length.fillna(1, inplace = True)
iris
```

```python
# Now let's delete the column class
del iris['class']
iris.head()
```

```python
# Set the first 3 rows as NaN
iris.iloc[0:3 ,:] = np.nan
iris.head()
```

```python
# Delete the rows that have NaN
iris = iris.dropna(how='any')
iris.head()
```

```python
# Reset the index so it begins with 0 again

iris = iris.reset_index(drop = True)
iris.head()
```
