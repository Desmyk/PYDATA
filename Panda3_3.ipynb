```python
# Import the necessary libraries

import numpy as np

import pandas as pd
```

```python
# Import the first dataset cars1 and cars2.
# assign variable cars1 and cars2

cars1 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv')

cars2 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv')

print(cars1)
```

```python
# Oops, it seems our first dataset has some unnamed blank columns, fix cars1

cars1 = cars1.loc[:, "mpg":"car"]
cars1.head()
```

```python
# What is the number of observations in each dataset?

observations_cars1 = cars1.shape[0]
observations_cars2 = cars2.shape[0]


print(f"Number of observations in cars1: {observations_cars1}")
print(f"Number of observations in cars2: {observations_cars2}")
```

```python
# Join cars1 and cars2 into a single DataFrame called cars

cars = pd.concat([cars1, cars2])

print(cars)
```

```python
# Oops, there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

owners = np.random.randint(15000, high=73001, size=398, dtype='l')
print(owners)
```

```python
# Add the column owners to cars

cars['owners'] = owners
cars.tail()
```
