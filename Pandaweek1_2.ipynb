```python
#  Use Pandas Datarader to read the daily time series for the Apple stock (ticker symbol AAPL) between 01/01/2015 and today,
#assign it to df_apple and print it.

import pandas_datareader.data as web

api_key = "HFQZ62XLBEZQXFYT"

# Get data from Alpha Vantage
df_apple = web.DataReader("AAPL", data_source="av-daily", api_key=api_key, start=start_date, end=end_date)

df_apple
```

```python
# Repeat the two previous steps for a few other stocks, always creating a new dataframe:
#  Tesla, IBM and Microsoft. (Ticker symbols TSLA, IBM and MSFT.)


api_key = "HFQZ62XLBEZQXFYT"

df_tesla = web.DataReader("TSLA", data_source="av-daily", api_key=api_key, start=start_date, end=end_date)
df_tesla["stock"] = "TSLA"

df_ibm = web.DataReader("IBM", data_source="av-daily", api_key=api_key, start=start_date, end=end_date)
df_ibm["stock"] = "IBM"

df_microsoft = web.DataReader("MSFT", data_source="av-daily", api_key=api_key, start=start_date, end=end_date)
df_microsoft["stock"] = "MSFT"

df_tesla
df_ibm
df_microsoft
```

```python
# Combine the four separate dataFrames into one combined dataFrame df that holds the information for all four stocks

df = pd.concat([df_apple, df_tesla, df_ibm, df_microsoft])
df
```

```python
#  Shift the stock column into the index (making it a multi-level index consisting of the ticker symbol and the date).
df = df.reset_index()

df.set_index(["stock", "index"], inplace=True)

df
```

```python
# Create a dataFrame called vol, with the volume values.

vol = df["volume"]

vol
```

```python
# Aggregate the data of volume to weekly.
# Hint: Be careful to not sum data from the same week of 2015 and other years.

vol = df["volume"]

# Convert the index to datetime
vol.index = pd.to_datetime(vol.index)

# Resample the data weekly, taking the sum of the volume for each week.
vol_weekly = vol.resample("W").sum()

# Print the vol_weekly DataFrame
vol_weekly
```

```python
# Find all the volume traded in the year of 2015
vol.index = pd.to_datetime(vol.index)

vol_2015 = vol[vol.index.year == 2015]

# Calculate the total volume traded in 2015
total_volume_2015 = vol_2015.sum()

# Print the result
print(f"Total volume traded in 2015: {total_volume_2015}")
```
