```python
import pandas as pd
import numpy as np
```

```python
# import dataset
# assign variable crime
crime = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')

print(crime)
```

```python
# What is the type of the columns?

column = crime.dtypes

# Print the column types
print(column)
```

```python
# Convert the type of the column Year to datetime64

crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')

print(crime.dtypes)
```

```python
# Set the Year column as the index of the dataframe

crime = crime.set_index('Year')
crime['Year'] = crime.index
print(crime)
```

```python
# Delete the Total column

del crime['Total']

print(crime.head())
```

```python
# Group the year by decades and sum the values
crime['Year'] = crime['Year'].astype('int64')

crime = crime.groupby(crime['Year'] // 10 * 10).sum()

crime = crime.iloc[:, 1:]

print(crime)
```

```python
# What is the most dangerous decade to live in the US?
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')

violent_decade = crime.groupby(crime['Year'].dt.year)['Violent'].sum()

most_dangerous = violent_decade.idxmax()

print(f"The most dangerous decade to live in the US was the {most_dangerous}'s.")
```
